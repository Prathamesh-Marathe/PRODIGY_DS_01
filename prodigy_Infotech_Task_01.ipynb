```python
# Clean column names (Used to remove and leading or trailing white space)
df.columns = df.columns.str.strip()
```

```python
df.drop(['2024', 'Unnamed: 69'], axis=1, inplace=True)
```

```python
# List all year columns as strings
year_cols = [col for col in df.columns if col.isdigit() and 1960 <= int(col) <= 2025]
print(year_cols)
```

```python
# Rename year columns to integers
rename_dict = {col: int(col) for col in year_cols}
df.rename(columns=rename_dict, inplace=True)
```

```python
# Fill missing values in year columns with column mean
for col in rename_dict.values():
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
# Filter out aggregate country groups
countries_to_exclude = [
    "Low & middle income","Middle income","IBRD only","Early-demographic dividend","Upper middle income", "High income", "Low income",
    "Lower middle income", "Late-demographic dividend", "Post-demographic dividend", "IDA & IBRD total", "IDA total", "IDA blend",
    "IDA only", "Euro area", "East Asia & Pacific (excluding high income)","East Asia & Pacific","Europe & Central Asia (excluding high income)",
    "Europe & Central Asia","Latin America & Caribbean (excluding high income)","Latin America & Caribbean","Middle East & North Africa",
    "Middle East & North Africa (excluding high income)","North America","South Asia","Sub-Saharan Africa (excluding high income)","Sub-Saharan Africa",
    "World","Arab World", "Central Europe and the Baltics","Caribbean small states","Central African Republic", "Channel Islands",
    "Developing small island states","East Asia & Pacific (IDA & IBRD countries)","Europe & Central Asia (IDA & IBRD countries)",
    "Fragile and conflict affected situations","Heavily indebted poor countries (HIPC)","High income: nonOECD","High income: OECD",
    "Latin America & the Caribbean (IDA & IBRD countries)","Least developed countries: UN classification","Middle East & North Africa (IDA & IBRD countries)",
    "Pacific island small states","Small states","South Asia (IDA & IBRD)","Sub-Saharan Africa (IDA & IBRD countries)",
    "Other small states","Pre-demographic dividend","European Union","Least Developed Countries","Less Developed Regions",
    "More Developed Regions","Land Locked Developing Countries","Small Island Developing States","Eastern and South-Eastern Asia",
    "Northern Africa and Western Asia","Oceania","Northern America and Europe","Eastern and Southern Africa","Middle Africa",
    "Western Africa","Western Asia","OECD members","Not classified"
]

# Apply the filter to the DataFrame
df = df[~df["Country Name"].isin(countries_to_exclude)]
```

```python
#  Histogram: Population distribution in 2023
plt.figure(figsize=(12,6))
sns.histplot(df[2023], bins=50, kde=True)
plt.title('Population Distribution in 2023')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Barplot: Population over years for India
country = 'India'
years = sorted(rename_dict.values())  # From 1960 to 2023
population = df[df['Country Name'] == country][years].values.flatten()

plt.figure(figsize=(14,6))
sns.barplot(x=years, y=population, color='salmon')
plt.title(f'Population of {country} from 1960 to 2023')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
#  Histogram: Population Growth from 2000 to 2020
df['growth_2000_2020'] = df[2020] - df[2000]

plt.figure(figsize=(12,6))
sns.histplot(df['growth_2000_2020'], bins=50, kde=True, color='lightgreen')
plt.title('Population Growth from 2000 to 2020')
plt.xlabel('Growth in Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()
```

```python
#  Overlapping Histograms: 1960 vs 2020
plt.figure(figsize=(12,6))
sns.histplot(df[1960], bins=20, color='blue', label='1960', kde=True, alpha=0.5)
sns.histplot(df[2020], bins=20, color='red', label='2020', kde=True, alpha=0.5)
plt.title('Population Distribution in 1960 vs 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Top 10 Most Populated Countries in 2023

top_10 = df[['Country Name', 2023]].sort_values(by=2023, ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_10[2023], y=top_10['Country Name'], palette='viridis')
plt.title('Top 10 Most Populated Countries in 2023')
plt.xlabel('Population')
plt.ylabel('Country')
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
# Bottom 10 Least Populated Countries in 2023

bottom_10 = df[['Country Name', 2023]].sort_values(by=2023, ascending=True).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=bottom_10[2023], y=bottom_10['Country Name'], palette='coolwarm')
plt.title('Bottom 10 Least Populated Countries in 2023')
plt.xlabel('Population')
plt.ylabel('Country')
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
countries = ['India', 'China']
years = sorted(rename_dict.values())  # [1960, ..., 2023]

plt.figure(figsize=(14,6))

for country in countries:
    pop = df[df['Country Name'] == country][years].values.flatten()
    plt.plot(years, pop, label=country, linewidth=2)

plt.title('Population Growth: India vs China (1960–2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# % Growth from 1960 to 2023 for India and China

df_ind_china = df[df['Country Name'].isin(['India', 'China'])][[1960, 2023, 'Country Name']].copy()
df_ind_china['% Growth'] = ((df_ind_china[2023] - df_ind_china[1960]) / df_ind_china[1960]) * 100

plt.figure(figsize=(8,5))
sns.barplot(data=df_ind_china, x='Country Name', y='% Growth', palette='Set2')
plt.title('Population Growth (%) from 1960 to 2023')
plt.ylabel('Percent Growth')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
# Line Plot: Population Growth Over Time for Top 5 Most Populated Countries in 2023

top_countries = df[['Country Name', 2023]].sort_values(by=2023, ascending=False).head(5)['Country Name'].values
years = sorted(rename_dict.values())

plt.figure(figsize=(14,7))

for country in top_countries:
    pop = df[df['Country Name'] == country][years].values.flatten()
    plt.plot(years, pop, label=country)

plt.title('Population Growth Over Time (Top 5 Countries)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
